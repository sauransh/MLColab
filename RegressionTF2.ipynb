```python
filter_df['date'] = pd.to_datetime(df['date'])
```

```python
filter_df['year'] = filter_df['date'].apply(lambda date:date.year)
```

```python
filter_df['month'] = filter_df['date'].apply(lambda date:date.month)
```

```python
filter_df.drop('zipcode',axis=1,inplace=True)
```

```python
X = filter_df.drop('price',axis=1).values
y = filter_df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101,)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.head()
```

```python
losses.plot()
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)
```

```python
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
```
